```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df=pd.read_csv('C:\\Users\\T4440S\\OneDrive\\Documents\\telco_customer_churn_queried.csv')
```

```python
df
```

```python
#lets change the total0charges to total charges

df=df.rename(columns={'Total0Charges':'Total_Charges'})
```

```python
#i have 5174 null values under churn reason, lets check the full row to know where it is
df[df['Churn Reason'].isnull()]
```

```python
df=df.fillna({'Churn Reason':'Unknown'})
```

```python
df
```

```python
#lets check if null still exist
df.isnull().sum()
```

```python
#now lets check for churn distribution

df['Churn Label'].value_counts()
```

```python
#lets confirm all columns have right types

df.dtypes
```

```python
#lets drop some unecesary collumns

df.drop(columns=['Count','Lat Long', 'Latitude', 'Longitude'], inplace= True)
```

```python
#lets check if the colums we dropped are still there
df.columns
```

```python
#after dropping some columns, lets me check if there is any null values again

df.isnull().sum()
```

```python
#lets make 2 visuals matplotib and seaborn. first is churn distribution

sns.countplot(x='Churn Label',data=df)
plt.title('Churn Distribution')
plt.show
```

```python
#lets visualise churn by contract

sns.countplot(x='Contract', hue='Churn Label',data=df)
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.show()
```
